1. Synthesize a compound signal by creating 4 SinSignal and CosSignal objects and adding them up. They should be harmoneous. (ex, signal = (SinSignal(freq=400, amp=1.0) + SinSignal(freq=600, amp=0.5))

```python
from thinkdsp import CosSignal, SinSignal

#SinSignal 및 CosSignal 객체 생성
cos_sig_1 = CosSignal(freq=880, amp=0.5)
cos_sig_2 = CosSignal(freq=1760, amp=0.25)

sin_sig_1 = SinSignal(freq=440, amp=1.0)
sin_sig_2 = SinSignal(freq=1320, amp=0.25)

#더해서 합성 신호 만들기
mix = cos_sig_1 + cos_sig_2 + sin_sig_1 + sin_sig_2
```

2. Evaluate the signal to get a Wave, and listen to it. 

```python
#신호를 파형으로 만들기
wave = mix.make_wave(duration=1, start=0, framerate=11025)
from IPython.display import Audio
audio = Audio(data=wave.ys, rate=wave.framerate)
audio
```

3. Compute its Spectrum and plot it. 

```python
import matplotlib.pyplot as plt  

#스펙트럼 계산
spectrum = wave.make_spectrum()

#스펙트럼 그리기
spectrum.plot()

#그래프 제목과 축 라벨 추기
plt.title("Signal Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")

#그래프 그리기
plt.show()
```

4. Add frequency components that are not multiples of the fundamental. And listen to it. What's the difference between the sound in 2 and 4?

```python
#주파수 추가
sin_sig_3 = SinSignal(freq=200, amp=2.0)
mix_2 = mix + sin_sig_3
wave = mix_2.make_wave(duration=1, start=0, framerate=11025)
audio = Audio(data=wave.ys, rate=wave.framerate)
audio
```

비교: fundamental frequency의 배수가 아닌 주파수를 추가했을 때 소리가 탁해지고 불협화음 같다.
